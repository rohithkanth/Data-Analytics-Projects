```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
val = input("Enter the medicine you want substitutions for : ")
print("the Medicine you entered is : " +val)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)
driver.get('https://www.1mg.com')
search_bar = driver.find_element(by=By.NAME,value="name")
search_bar.clear()
search_bar.send_keys(val)
search_bar.send_keys(Keys.RETURN)

def extract_medicine_title(driver):

    medicine_list= []
    elems = driver.find_elements(by=By.XPATH,value='//span[@class ="style__pro-title___3zxNC"]')

    for p in range(len(elems)):
            medicine_list.append(elems[p].text)
    if medicine_list == []:
        print('Not found')
    else:

        return(medicine_list)
medicine_title = extract_medicine_title(driver)

def extract_medicine_links(driver):
    links = []
    li = driver.find_elements(By.XPATH,'//*[@class = "col-xs-12 style__container___cTDz0"]/div[1]/a')
    for i in li:
        links.append(i.get_attribute('href'))
    if links == []:
        print('Not found')
    else:
        return(links)
links = extract_medicine_links(driver)
driver.get(links[0])

def extract_medicine_manufacturer(driver):
    manu= driver.find_element(by=By.XPATH,value='//div[@class ="DrugHeader__meta-value___vqYM0"]')
    return(manu.text)
search_company = extract_medicine_manufacturer(driver)

def extract_salt_composition(driver):
    salt= driver.find_element(by=By.XPATH,value='//div[@class ="saltInfo DrugHeader__meta-value___vqYM0"]')
    return(salt.text)
salt_name=extract_salt_composition(driver)

print("Searching alternatives for " + "\033[1m"  +medicine_title[0]+ "\033[0m" + " manufacturer " "\033[1m" +search_company+ "\033[0m" + " having salt composition " + "\033[1m" +salt_name )
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)
driver.get('https://www.1mg.com')
search_bar = driver.find_element(by=By.NAME,value="name")
search_bar.clear()
search_bar.send_keys(salt_name)
search_bar.send_keys(Keys.RETURN)

alternate_title = extract_medicine_title(driver)
alternate_link = extract_medicine_links(driver)
def extract_drug_header(driver):

    drug_header = driver.find_element(by=By.XPATH,value ="//h1[@class='DrugHeader__title-content___2ZaPo']")
    return((drug_header.text))
alternate_manufacturer = []
alternate_drug_header = []
for i in range(5):
     drug_header=extract_drug_header(driver)
    manufacturer = extract_medicine_manufacturer(driver)
    alternate_manufacturer.append(manufacturer)
    alternate_drug_header.append(drug_header)

df_alternate_medicine = pd.DataFrame({"Alternate Medicine": alternate_drug_header , "Alternate Medicine Company" : alternate_manufacturer})
df_alternate_medicine
```
